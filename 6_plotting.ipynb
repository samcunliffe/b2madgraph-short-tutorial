```python
import uproot
import matplotlib.pyplot as pp
```

```python
fg = uproot.open("gen_tuple.root")
beam = fg["beam"]
```

```python
pp.hist(beam['alp_InvM'].array(), bins=100)
pp.xlabel(r"$m_{a}$ [GeV/$c2$]")
pp.ylabel("gen-level events")
pp.show()
```

```python
fr = uproot.open("analysis_tuple.root")
reco = fr['reco']
```

```python
pp.hist(reco['daughter__bo0__cm__spInvM__bc'].array(), bins=100)
pp.xlabel(r"$M_{\gamma\gamma}$ [GeV/$c^2$]")
pp.ylabel("reconstructed candidates")
pp.show()
```
